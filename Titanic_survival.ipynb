Distributions

```python
histogram_labels = ['Fare',
                    'Age'
                   ]

colors = px.colors.qualitative.Vivid
for i in range(len(histogram_labels)):
    fig = px.histogram(df,
                       histogram_labels[i],
                       title=f'{histogram_labels[i]} distribution in top hits',
                       color_discrete_sequence=[colors[i]]
                      )
    iplot(fig)
```

Titanic split

```python
df.columns
```

```python
df.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

```python
df
```

```python
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
```

```python
df
```

```python
X = df.drop(columns='Survived',axis=1)
Y = df['Survived']
```

```python
X
```

```python
Y
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
X_train
```

```python
x_test
```

```python
Y_train
```

```python
y_test
```

Visualization of the realtion between the columns using corr

```python
corrcolumns = df.drop(columns=['Embarked','Sex'])
corrcolumns
```

```python
sns.heatmap(corrcolumns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
```

Logistic Regression Model

```python
logmodel = LogisticRegression(max_iter=5000,  C=0.1)
```

```python
logmodel.fit(X_train,Y_train)
```

```python
y_pred = logmodel.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
logmodel.score(x_test, y_test)
```

```python
logmodel.score(X_train, Y_train)
```

Model Evaluations

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
